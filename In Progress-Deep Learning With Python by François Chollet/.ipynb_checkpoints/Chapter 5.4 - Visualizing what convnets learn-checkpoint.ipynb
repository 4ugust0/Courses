## Example data preprocessing

```python
# Path
img_path = 'E:/1_GitHub/arcyfelix/Courses/In Progress-Deep Learning With Python by Fran√ßois Chollet/data/Chapter 5.2 - Using convets with small datasets/test/cats/cat.1752.jpg'
```

```python
from keras.preprocessing import image
import numpy as np
```

### Preprocessing the image into a 4D tensor

```python

img = image.load_img(img_path, 
                     target_size = (150, 150))

img_tensor = image.img_to_array(img)
```

```python
img_tensor.shape
```

```python
img_tensor = np.expand_dims(img_tensor, 
                            axis = 0)
```

```python
img_tensor.shape
```

```python
# Remember that the model was trained on inputs
# that were preprocessed in the following way:
img_tensor /= 255.
```

```python
import matplotlib.pyplot as plt

plt.imshow(img_tensor[0])
plt.show()
```
