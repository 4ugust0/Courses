# Chapter 3.4.1 - Classifying movie reviews The IMDB dataset,

```python
# Loading the IMDB dataset
from keras.datasets import imdb
```

```python
# Chosing number of the most frequent word to be kept. 
num_words = 10000
```

```python
(train_data, train_labels), (test_data, test_labels) = imdb.load_data(num_words = num_words)
```

```python
# Showing the first example
train_data[0]
```

```python
# Label of the first review
train_labels[0]
```

```python
# Getting dictionary mapping words to an integer index
word_index = imdb.get_word_index()
```

```python
word_index
```

```python
# Reversing the word index to get mapping from indexes to words
reverse_word_index = dict([(value, key) for (key,value) in word_index.items()])
```

```python
reverse_word_index
```

```python
# Decoding the review
# First 3 indexes are reserved for padding (0), start of the sequence (1) and unknown (2) 
decoded_review = ' '.join([reverse_word_index.get(i - 3, '?') for i in train_data[0]])
```

```python
decoded_review
```

## Preparing the data by encoding the integer sequence into a binary matrix

```python
import numpy as np
```

```python
def vectorize_sequences(sequences, dimension = 10000):
    # Create an all-zero matrix of shape (len(sequences), dimension)
    results = np.zeros((len(sequences), dimension))
    for i, sequence in enumerate(sequences):
        # Setting specific indices of results[i] to 1s
        results[i, sequence] = 1.  
    return results
```

```python
train_data.shape
```

```python
x_train = vectorize_sequences(train_data)
```

```python
x_train.shape
```

```python
x_test = vectorize_sequences(test_data)
```

```python
# Prompting the first example
x_train[0]
```

```python
# Vectorizing the labels
y_train = np.asarray(train_labels)
y_train = y_train.astype('float32')
```

```python
y_train
```

```python
y_train.shape
```

```python
y_test = np.asarray(test_labels)
y_test = y_test.astype('float32')
```

## Building the neural network

```python
# Improting building block of the network
from keras.models import Sequential
from keras.layers import Dense
```

```python
# Creating the network
model = Sequential()
model.add(Dense(units = 16, 
                activation = 'relu', 
                input_shape = (10000,)))
model.add(Dense(units = 16, 
                activation = 'relu'))
# Output layer
# It is a binary classification so we can use sigmoid activation function
model.add(Dense(units = 1, 
                activation = 'sigmoid'))
```

```python
# Compiling the network with a custom hyperparameters
from keras import optimizers
from keras import losses
from keras import metrics

model.compile(optimizer='rmsprop',
              loss='binary_crossentropy',
              metrics=['accuracy'])
```

```python
model.summary()
```

## Setting aside a validation set

```python
x_val = x_train[:10000]
partial_x_train = x_train[10000:]

y_val = y_train[:10000]
partial_y_train = y_train[10000:]
```

## Training the model

```python
history = model.fit(partial_x_train, 
                    partial_y_train, epochs = 20, 
                    batch_size = 512,
                    validation_data = (x_val, y_val))
```

```python
history_dict = history.history
history_dict.keys()
```

```python
import matplotlib.pyplot as plt

acc = history.history['acc']
val_acc = history.history['val_acc']
loss = history.history['loss']
val_loss = history.history['val_loss']

epochs = range(1, len(acc) + 1)

# "bo" is for "blue dot"
plt.plot(epochs, loss, 'bo', label='Training loss')
# b is for "solid blue line"
plt.plot(epochs, val_loss, 'b', label='Validation loss')
plt.title('Training and validation loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()

plt.show()
```

```python
# Clearing the figure
plt.clf()  
acc_values = history_dict['acc']
val_acc_values = history_dict['val_acc']

# bo == blue dot
plt.plot(epochs, acc, 'bo', label='Training acc')
# b == blue line (line is default)
plt.plot(epochs, val_acc, 'b', label='Validation acc')
plt.title('Training and validation accuracy')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()

plt.show()
```

```python
results = model.evaluate(x_test, y_test)
```

```python
results
```

## Retraining the model

```python
# The model obtains the best accuracy for epoch 4 so let's retrain it 
# After the 4th epoch, the network overfits.
model = Sequential()
model.add(Dense(units = 16, 
                activation = 'relu', 
                input_shape = (10000,)))
model.add(Dense(units = 16, 
                activation = 'relu'))
model.add(Dense(units = 1, 
                activation = 'sigmoid'))
model.compile(optimizer = 'rmsprop',
              loss = 'binary_crossentropy',
              metrics = ['accuracy'])
model.fit(x_train, 
          y_train, 
          epochs = 4, 
          batch_size = 512)
```

```python
results = model.evaluate(x_test, y_test)
```
